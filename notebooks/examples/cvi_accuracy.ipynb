```
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_validate

from gamac.pipeline.cvi_predictor import load_pickle

meta_classifier = load_pickle('classifier.pkl')
meta_df = pd.read_csv('../data/meta-dataset.csv')

print(meta_df.head())

meta_values = meta_df.values
meta_features, meta_labels = meta_values[:, :-1], meta_values[:, -1]

uniq_measures = np.unique(meta_labels)
print("Unique measures: ", uniq_measures)

measures_idx = {m: idx for idx, m in enumerate(uniq_measures)}
measures = [measures_idx[m] for m in meta_labels]

results = cross_validate(
    meta_classifier,
    meta_features,
    measures,
    cv=20,
    scoring='f1_macro'
)

score = np.mean(results['test_score'])
print(f"F1 score: {score}")
```
